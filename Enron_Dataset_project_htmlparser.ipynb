```python
#extracting a two-person chat conversation
##to-do; implement a three or more person chat extraction

import os, re

#!pip install lxml



#!pip install beautifulsoup4


import urllib
from bs4 import BeautifulSoup 

all_data=[]

file_list = glob.glob('*.html') # returns list of files
for file in file_list:
    with open(file,'rb') as fp:
        soup = BeautifulSoup(fp)
        
        for script in soup(["script", "style"]):
            script.extract()    # rip it out
            
            # get text
            text = soup.get_text()

            # break into lines and remove leading and trailing space on each
            lines = (line.strip() for line in text.splitlines())
            # break multi-headlines into a line each
            chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
            # drop blank lines
            text = '\n'.join(chunk for chunk in chunks if chunk)

            #get number of rows (ensure its even, if not add empty row to dataframe)
            text=text.split("\n")
            rows= len(text)
            mod_rows=(text.append('')) if (rows % 2) else rows
            rows_num= len(text)
            #string_rows = [str(integer) for integer in mod_rows]
            #a_string = "".join(string_rows)
            #row_integer = int(a_string)
            
            df = pd.DataFrame(np.array(text).reshape(int(rows_num/2),2), columns = ['Person','Message'])
            #df = df.iloc[2:]
            part= [df.loc[1:1]]
            df['Person'] = df['Person'].str.extract('([a-zA-Z ]+)', expand=False).str.strip()
    
            all_data.append(df)
```
